### Visualizing the Decision Tree Classifier using Iris data

```python
# Importing the libraries
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
# Loading the data
iris = load_iris()
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df['Target'] = iris.target
```

```python
# Displaying the first 5 rows
df.head()
```

```python
# Displaying the last 5 rows
df.tail()
```

### Defining the Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
```

```python
# Defining the dependent and independent variables
X = df.drop('Target',axis =1)
y = df['Target']
```

```python
# Applying the decision tree classifier algorithm on the data
dtc.fit(X,y)
```

### Visualizing the Decision Tree Classifier

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(20,10))
dtplot = plot_tree(dtc, feature_names = iris.feature_names, class_names = iris.target_names, filled = True)
```

### Predicting on the sample data

```python
prediction1 = dtc.predict([[4.2,3.2,1.2,0.4]])
prediction1
```
